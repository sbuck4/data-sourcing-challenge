### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
built_url = (
    f"{url}api-key={nyt_api_key}&begin_date={begin_date}&end_date={end_date}"
    + f'&fq={filter_query}&sort={sort}&fl={field_list}'
)

# query_url
```

```python
# Create an empty list to store the reviews
reviews = []

# loop through pages 0-19
for page in range(20):

    # create query with a page number
    # API results show 10 articles at a time
    query_url = f'{built_url}?page={page}'
    
    # Make a "GET" request and retrieve the JSON
    response = requests.get(query_url)
    json_data = response.json()

    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(2)
    
    # Try and save the reviews to the reviews_list
    if 'response' in json_data and 'docs' in json_data['response']:

        # loop through the reviews["response"]["docs"] and append each review to the list
        for doc in json_data['response']['docs']:
            reviews.append(doc)
        
        # Print the page that was just retrieved
        print(f"Successfully retrieved page {page}")
        
    else:

        # Print the page number that had no results then break from the loop
        print(f"No results on page {page}")
        break

    print(f"Total reviews collected: {len(reviews)}")
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
preview_reviews = reviews[:5]
print(json.dumps(preview_reviews, indent=4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
from pandas import json_normalize

df = json_normalize(reviews)
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early
# if 'headline.main' in df.columns:
if 'headline.main' in df.columns:
    
    def extract_title(headline):
        if pd.isnull(headline):
            return None
        start = headline.find('\u2018') + 1
        end = headline.find('\u2019') + len(" Review")
        title = headline[start:end]
        return title.strip()

    df['title'] = df['headline.main'].apply(extract_title)

    print(df[['headline.main', 'title']].head())
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:

        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 

        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
if 'keywords' in df.columns:
    df['keywords'] = df['keywords'].apply(extract_keywords)

    print(df[['keywords']].head())
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
```

```python
# Create an empty list to store the results
results = []

# Create a request counter to sleep the requests after a multiple
# of 50 requests
request_counter = 0
max_requests_before_sleep = 50

# Loop through the titles
for title in title_list:

    # Check if we need to sleep before making a request
    if request_counter >= max_requests_before_sleep:
        print("Sleeping for 10 seconds to avoid hitting rate limits")
        time.sleep(10)  
        request_counter = 0  

    # Add 1 to the request counter
    request_counter += 1
    
   # Perform a "GET" request for The Movie Database
    try:
        response = requests.get(f"{url}{title}{tmdb_key_string}")
        response.raise_for_status()  # Raise an exception for HTTP error responses

        # Process the response JSON
        data = response.json()
        if data['results']:
            # Get the first movie result
            movie = data['results'][0]

            # Get movie id
            movie_id = movie['id']

            # Make a request for the full movie details
            movie_details_url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb_api_key}"
            
            # Execute "GET" request with url
            movie_details_response = requests.get(movie_details_url)
            movie_details_response.raise_for_status()
            movie_details = movie_details_response.json()

            # Extract the genre names into a list
            genres = [genre['name'] for genre in movie_details.get('genres', [])]

            # Extract the spoken_languages' English name into a list
            spoken_languages = [language['english_name'] for language in movie_details.get('spoken_languages', [])]

            # Extract the production_countries' name into a list
            production_countries = [country['name'] for country in movie_details.get('production_countries', [])]

            # Add the relevant data to a dictionary and append it to the results list
            movie_data = {
                'title': movie_details.get('title', 'N/A'),
                'release_date': movie_details.get('release_date', 'N/A'),
                'runtime': movie_details.get('runtime', 'N/A'),
                'genres': genres,
                'spoken_languages': spoken_languages,
                'production_countries': production_countries,
                'overview': movie_details.get('overview', 'N/A')
            }
            results.append(movie_data)

            # Print out the title that was found
            print(f"Found movie: {movie_data['title']}")

        else:
            print(f"Movie not found for title: {title}")

    except requests.RequestException as e:
        # Handle HTTP errors and other request exceptions
        print(f"Request failed for title: {title}. Error: {e}")
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
for result in results[:5]:
    formatted_result = json.dumps(result, indent=4)
    print(formatted_result)
```

```python
# Convert the results to a DataFrame
df_results = pd.DataFrame(results)
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merged_df = pd.merge(df, df_results, on='title', how='inner')

print(merged_df.columns)
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
columns_to_fix = ['genres', 'spoken_languages_x', 'spoken_languages_y', 'production_countries_x', 'production_countries_y']

# Create a list of characters to remove
characters_to_remove = ['[', ']', "'", '"']

# Loop through the list of columns to fix
for col in columns_to_fix:
    # Convert the column to type 'str'
    merged_df[col] = merged_df[col].astype(str)
    
    # Loop through characters to remove
    for char in characters_to_remove:
        merged_df[col] = merged_df[col].str.replace(char, '', regex=False)

# Display the fixed DataFrame
print(merged_df.head())

```

```python
# Drop "byline" column
merged_df = merged_df.drop(columns=['byline'])
print(merged_df.head())
```

```python
# Delete duplicate rows
merged_df = merged_df.drop_duplicates()

# Reset index after removing duplicates
merged_df = merged_df.reset_index(drop=True)

# Display the DataFrame to verify changes
print(merged_df.head())
```

```python
# Export data to CSV without the index
merged_df.to_csv('merged_data.csv', index=False)

print("Data exported to 'merged_data.csv' without the index.")
```
